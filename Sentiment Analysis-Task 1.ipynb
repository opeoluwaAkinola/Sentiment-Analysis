```python
# importing Useful Modules
import requests, urllib
# import http.client
from pathlib import Path
# from datetime import datetime
import pandas as pd
from bs4 import BeautifulSoup
```

Create directory for raw data storage, if it does not already exist:

```python
# directory for raw data storage
dir_raw = Path("raw_data")
# make sure it exists
dir_raw.mkdir(parents=True, exist_ok=True)
```

```python
# Max pages in for each month of the year
max_pages=7
#columns of review data
columns=["Title","Star Rating","Review","Helpfulness"]
#months of the year
months=["jan","feb","mar","apr","may","jun","jul","aug","sep","oct","nov","dec"]
#year range of the reviews
years=["2016","2017","2018","2019","2020","2021"]
#root of the URL
URLroot="http://mlg.ucd.ie/modules/python/assign2/18201236/reviews"
```

# Data Collection
I created a function <b>getreviews</b> to search retrieve reviews from the website pages form each month of the different years.

```python
def getreviews(month,year):
    #row to store each review document
    rows=[]
    #loop through the pages for specified year and month
    for i in range(max_pages):
        #exception to break once review data in all pages are retrieved
        try:
            #page request
            page=requests.get(URLroot+"-"+year+"-"+month+"-0"+str(i+1)+".html")
            #Initialize beautiful soup object to retrieve page content
            soup = BeautifulSoup(page.content, 'html.parser')
            #create a method to retrieve the review data.This is because the tags for the revies are not all the same
            #Some have attributes of class:"review" while the others have class:"review-alt".
            #Method 'reviews' makes the code shorter and simpler and takes in the specified review tag and the rows dict
            def reviews (listin,rows):
                #loop through all found review tags
                for review in listin:
                    row={}
                    #get the title of the reviews, titles either have class attribute 'review-title' or 'review-title2'
                    if review.find("h5",attrs={'class':'review-title'}):
                        a=review.find("h5",attrs={'class':'review-title'})
                    else:
                        a=review.find("h5",attrs={'class':'review-title2'})
                    #add exceptions to by=pass 'attributeError' in beautifulsoup objects.
                    try:
                        row[columns[0]]=a.text
                    #Get star rating, rating is only specified in <span> tag,got rating using string handling
                        for n in a.children:
                            d=str(n)
                            break
                        row[columns[1]]=d.replace('<span class="stars"><img alt="', '')[0]
                    except:
                        pass
                    #Get body of the review
                    row[columns[2]]=review.find("p",attrs={'class':'review-body'}).text
                    #Get Review helpfullness
                    row[columns[3]]=review.findAll("p",attrs={'class':'metadata'})[1].text
                    rows.append(row)
                    count=+1
                return rows
            #call review mthod with different review tags
            rows=reviews(soup.findAll('div',attrs={'class':'review'}),rows)
            rows=reviews(soup.findAll('div',attrs={'class':'review-alt'}),rows)
        except HTTPError:
            break    
    print(str(len(rows)) +" reviews found for "+month+" "+year) 
    return pd.DataFrame(rows)
```

I created a second function <b>savereviews</b> to save data to csv file

```python
def savereviews(month,year):
    df=getreviews(month,year)
    # write it out to our raw dataset directory
    fname = "%s-%s-reviews.csv" % (month,year)
    out_path = dir_raw/fname
    print("Writing data to %s/%s" % (out_path,fname))
    fout = open(out_path, "w")
    df.to_csv(out_path,index=False)
    fout.close()
```

```python
for year in years:
    for month in months:
        savereviews(month,year)
```
